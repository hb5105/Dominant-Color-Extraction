# Dominant Color Extraction

```python
#importing the neccessary libraries
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('elephant.jpg')
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
#converting BGR image into RGB
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
print(img2.shape)
```

```python
#flatten each channel of the image in a linear array
pixels=img2.reshape((330*500,3))
print(pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
k_dominant_colors=4
km=KMeans(n_clusters=k_dominant_colors)
km.fit(pixels)
```

```python
centers=km.cluster_centers_
centers=np.array(centers,dtype='uint8')
print(centers)
```

## Plotting these color values

```python
i=1
colors=[]
for each_color in centers:
    plt.subplot(1,4,i)
    i=i+1
    
    colors.append(each_color)
    
    #assign pixel value to corresponding color values
    color_graph= np.zeros((100,100,3),dtype='uint8')
    color_graph[:,:,:]=each_color
    plt.imshow(color_graph)
plt.show()
```

## Segmenting the original image

```python
img_new=np.zeros((330*500,3),dtype='uint8')
print(img_new.shape)
```

```python
label=km.labels_
print(len(label))
```

```python
#finding which color/label each pixel belongs to 

for i in range(img_new.shape[0]):
    img_new[i]=colors[km.labels_[i]]
    
img_new=img_new.reshape((img2.shape))


#new final image
plt.imshow(img_new)
plt.show()
```
