# problems with k means algo
## can fail for complex datasets
### it can fail if the data is not linearly seperable


```python
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_moons

x,y= make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import KMeans

km=KMeans(n_clusters=2)
km.fit(x)

centers=km.cluster_centers_
labels=km.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],color='red',marker='*')
```

```python
#FOR THIS WE HAVE DBSCAN, density based spatial clustering applications with noise
```

# DBSCAN
### based on the criteria that a cluster is a high density area


```python
from sklearn.cluster import DBSCAN
```

```python
#create a dbscan object
dbs=DBSCAN(eps=0.22,min_samples=5)
dbs.fit(x)
```

```python
ypred=dbs.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=ypred)
```
